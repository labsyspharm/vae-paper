```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
out = os.path.join(os.getcwd(), 'output/cluster_frequencies')
if not os.path.exists(out):
    os.makedirs(out)

```

```python
plt.rcParams['font.family'] = 'Arial'
```

```python
# Leiden cluster frequency
leiden = pd.read_csv(os.path.join(os.getcwd(), 'input/CRC-097_leiden.csv'))

fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(
    x=leiden.groupby('Cluster').size().sort_values().index.astype('str'),
    height=(leiden.groupby('Cluster').size().sort_values() / len(leiden)) * 100, color='k'
)
plt.xlabel('Leiden cluster')
plt.ylabel('Percent total cells')
plt.tight_layout()
plt.show()
fig.savefig(os.path.join(out, 'Leiden_cluster_frequency.pdf'))
plt.close('all')
```

```python
# HDBSCAN cluster frequency
hdbscan = pd.read_csv(os.path.join(os.getcwd(), 'input/CRC-097_hdbscan.csv'))
hdbscan = hdbscan[hdbscan['hdbscan'] != -1]

fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(
    x=hdbscan.groupby('hdbscan').size().sort_values().index.astype('str'),
    height=(hdbscan.groupby('hdbscan').size().sort_values() / len(hdbscan)) * 100, color='k'
)
plt.xlabel('HDBSCAN cluster')
plt.ylabel('Percent total cells')
plt.tight_layout()
plt.show()
fig.savefig(os.path.join(out, 'HDBSCAN_cluster_frequency.pdf'))
plt.close('all')
```
