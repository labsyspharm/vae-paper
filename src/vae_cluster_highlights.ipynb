```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
clustering = 'VAE9_ROT_VIG18'
```

```python
# Paths and input
out = os.path.join(os.getcwd(), f'output/cluster_hightlights/{clustering}')
if not os.path.exists(out):
    os.makedirs(out)

# Read single-cell sample for VAE analysis
main = pd.read_csv(os.path.join(os.getcwd(), 'input/main.csv'))

# REMOVE WHEN DONE
# df = pd.read_csv('/Users/greg/Downloads/vae_latent_vectors-cells (10).csv')
# main['test'] = df['Cluster']
```

```python
for i in sorted(main[clustering].unique()):
    if i != -1:
        print(f'{clustering} cluster {i}')
        plt.scatter(main[f'{clustering}_emb1'], main[f'{clustering}_emb2'], lw=0.0, s=0.01, c='gainsboro')
        plt.scatter(
            main[f'{clustering}_emb1'][main[clustering] == i],
            main[f'{clustering}_emb2'][main[clustering] == i],
            s=0.01, lw=0.0, c='tab:red'
        )
        plt.savefig(os.path.join(out, f'{i}.png'), dpi=800, bbox_inches='tight')
        plt.show()
        plt.close('all')
```
